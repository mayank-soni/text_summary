```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", model='cross-encoder/nli-distilroberta-base')
labels = ['sport', 'tech', 'finance', 'politics', 'business', 'economics', 'entertainment', 
          'beauty', 'travel', 'music', 'food', 'science', 'gaming', 'energy']
```

```python
from datasets import load_dataset

dataset = load_dataset("cnn_dailymail", '3.0.0')
```

```python
import pandas as pd
lst_dics = [dic for dic in dataset["test"]]

df = pd.DataFrame(lst_dics).rename(columns={"article":"text", 
      "highlights":"y"})[["text","y"]].head(20)
df
```

```python
df = df.iloc[:20]
df
```

```python
result = []

for i, text in enumerate(df['text']):
  print(f'loop no. {i}')
  result.append(classifier(text, labels))
```

```python
for subresult in result:
  print(subresult['labels'][0])
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline

bart_model = AutoModelForSequenceClassification.from_pretrained('navteca/bart-large-mnli')
bart_tokenizer = AutoTokenizer.from_pretrained('navteca/bart-large-mnli')
```

```python
#nlp = pipeline('zero-shot-classification', model=bart_model, tokenizer=bart_tokenizer)
```

```python
import torch
# for subtext in df['text']:
# features = bart_tokenizer(df['text'], padding=True, truncation=True, return_tensors="pt")
features = bart_tokenizer(dataset['test']['article'][10926], padding=True, truncation=True, return_tensors="pt")

bart_model.eval()
with torch.no_grad():
    scores = bart_model(**features).logits
    #label_mapping = ['sport', 'tech', 'finance', 'politics', 'business', 'economics', 'entertainment', 
                     #'beauty', 'travel', 'music', 'food', 'science', 'gaming', 'energy']
    label_mapping = ['sport', 'others']
    labels = [label_mapping[score_max] for score_max in scores.argmax(dim=1)]
    print(labels)
```
