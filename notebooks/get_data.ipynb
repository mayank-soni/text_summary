```python
import requests
import os
import pandas as pd
```

```python
def fetch_news():
    params = {'countries':'SG', 'lang':'en','sources':'channelnewsasia.com', 'from':'2022/11/10', 'to':'2022/11/20'}
    url = 'https://api.newscatcherapi.com/V2/search'
    r = requests.get(url, params=params)
    r.json()
    
    return r['articles']
```

```python
a = fetch_news()
```

```python
def get_headlines():
    url = "https://api.newscatcherapi.com/v2/latest_headlines"
    params = {'sources':'channelnewsasia.com'}
    headers = {
        "x-api-key": os.environ['NEWSCATCHER_API_KEY']
        }
    response = requests.get(url, headers=headers, params=params)
    return response.json()
```

```python
headline_json = get_headlines()
```

```python
df = pd.DataFrame(headline_json['articles'])
```

```python
def get_news_sources():
    params = {"countries":"SG","lang":"en"}
    url = 'https://api.newscatcherapi.com/v2/sources'
    headers = {"x-api-key": os.environ["NEWSCATCHER_API_KEY"]}
    r = requests.get(url, params=params, headers = headers)    
    return r.json()
```
