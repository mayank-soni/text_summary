```python
import requests
import os
import pandas as pd
```

## Newscatcher -- Request quota already hit

```python
def get_headlines():
    url = "https://api.newscatcherapi.com/v2/latest_headlines"
    params = {'sources':'channelnewsasia.com'}
    headers = {
        "x-api-key": os.environ['NEWSCATCHER_API_KEY']
        }
    response = requests.get(url, headers=headers, params=params)
    return response.json()
```

```python
headline_json = get_headlines()
headline_json
```

```python
df = pd.DataFrame(headline_json['articles'])
```

```python
df.columns
```

```python
links = []
summaries = []
excerpts = []
for i in range(5):
    links.append(df.loc[i, 'link'])
    summaries.append(df.loc[i, 'summary'])
    excerpts.append(df.loc[i, 'excerpt'])
```

```python
summaries
```

```python
excerpts
```

```python
def get_news_sources():
    params = {"countries":"SG","lang":"en"}
    url = 'https://api.newscatcherapi.com/v2/sources'
    headers = {"x-api-key": os.environ["NEWSCATCHER_API_KEY"]}
    r = requests.get(url, params=params, headers = headers)    
    return r.json()
```

## Newsapi - 100 requests per day

```python
# historical news can only access up to 1 month old
# 100 requests per day

def get_headlines2():
    url = "https://newsapi.org/v2/top-headlines"
    params = {"apiKey": os.environ['newsapi_API_KEY'], 'category':'sports', 'language':'en', 'from':'2022-11-01'}
    response = requests.get(url, params=params)
    return response.json()
```

```python
headline_json2 = get_headlines2()
headline_json2
```

```python
df2 = pd.DataFrame(headline_json2['articles'])
df2.head()
```

```python
links2 = []
titles2 = []
for i in range(10):
    links2.append(df2.loc[i, 'url'])
    titles2.append(df2.loc[i, 'title'])
```

```python
links2
```

```python
titles2
```

## Mediastack - 500 requests per month

```python
# limit to 500 requests per month

def get_headlines3():
    url = "http://api.mediastack.com/v1/news"
    params = {"access_key": os.environ['mediastack_API_KEY'], 'categories':'sports', 'languages':'en', 'date':'2022-10-01,2022-10-31'}
    response = requests.get(url, params=params)
    return response.json()
```

```python
headline_json3 = get_headlines3()
headline_json3
```

```python
df3 = pd.DataFrame(headline_json3['data'])
df3.head()
```

```python
links3 = []
titles3 = []
for i in range(10):
    links3.append(df3.loc[i, 'url'])
    titles3.append(df3.loc[i, 'title'])
```
